```python
import pandas as pd
import folium
import json
import geopandas as gpd
```

Import Data and set the map to the middle of Pittsburgh.
Load in all spots and chech id they are at the right spot.
Use https://fontawesome.com/ to change the icon of each spots. In this case, I want to use the "food" sign

```python
url = 'https://raw.githubusercontent.com/w94090/INFSCI2415-Pittsburgh-Fish-Fry-Map/main/2023Pittsburgh%20Fish%20Fry.csv'
df = pd.read_csv(url)
containers_map = folium.Map(location=[40.4406, -79.9959], zoom_start=12)
records = df[['longitude', 'latitude']].to_records()

containers_map = containers_map = folium.Map(location=[40.4406, -79.9959], zoom_start=11)
# Add the markers to the map...
for record in records:
    folium.Marker(location=[record[2], record[1]],
                  icon=folium.Icon(color='green',
                                   icon_color='white',
                                   icon='utensils',
                                   prefix='fa',
                                   )
                  ).add_to(containers_map)

containers_map
```

Nice! Then lets add tooltip so when we hover our mouse on the spot, we know what the name of the place is!

```python
url = 'https://raw.githubusercontent.com/w94090/INFSCI2415-Pittsburgh-Fish-Fry-Map/main/2023Pittsburgh%20Fish%20Fry.csv'
df = pd.read_csv(url)
containers_map = folium.Map(location=[40.4406, -79.9959], zoom_start=12)
records = df[['longitude', 'latitude','venue_name']].to_records()

containers_map = containers_map = folium.Map(location=[40.4406, -79.9959], zoom_start=11)
# Add the markers to the map...
for record in records:
    folium.Marker(location=[record[2], record[1]],
                  tooltip=record[3],
                  icon=folium.Icon(color='green',
                                   icon_color='white',
                                   icon='utensils',
                                   prefix='fa',
                                   )
                  ).add_to(containers_map)

containers_map
```

As we know from previous data exploration, we know that these venues can be held at Church, Fire Department, Restaurant and othe unsure places. Let's use color and icon to seperate them.

```python
url = 'https://raw.githubusercontent.com/w94090/INFSCI2415-Pittsburgh-Fish-Fry-Map/main/2023Pittsburgh%20Fish%20Fry.csv'
df = pd.read_csv(url)
containers_map = folium.Map(location=[40.4406, -79.9959], zoom_start=12)
color_dict = {'Church': 'brown',
              'Fire Department': 'red',
              'Restaurant': 'green',
              'Community Organization':'blue',
              "Veteran's Organization": 'orange',
              'Unsure / N/A': 'black',
              'Other':'gray',
              'Food Truck': 'darkpurple',
              'VFW': 'orange',
              'Market': 'lightgreen'
              }
records = df[['longitude', 'latitude','venue_name', 'venue_type','venue_address']].to_records()

containers_map = containers_map = folium.Map(location=[40.4406, -79.9959], zoom_start=11)
# Add the markers to the map...
for record in records:
    folium.Marker(location=[record[2], record[1]],
                  tooltip=record[3],
                  popup=record[5],
                  icon=folium.Icon(color=color_dict[record[4]],
                                   icon_color='white',
                                   icon='utensils',
                                   prefix='fa',
                                   )
                  ).add_to(containers_map)

containers_map
```

```python
url = 'https://raw.githubusercontent.com/w94090/INFSCI2415-Pittsburgh-Fish-Fry-Map/main/2023Pittsburgh%20Fish%20Fry.csv'
df = pd.read_csv(url)
containers_map = folium.Map(location=[40.4406, -79.9959], zoom_start=12)
color_dict = {'Church': 'beige',
              'Fire Department': 'red',
              'Restaurant': 'green',
              'Community Organization':'blue',
              "Veteran's Organization": 'orange',
              'Unsure / N/A': 'black',
              'Other':'gray',
              'Food Truck': 'darkpurple',
              'VFW': 'orange',
              'Market': 'darkgreen'
              }
icon_dict = {'Church': 'church',
              'Fire Department': 'fire',
              'Restaurant': 'utensils',
              'Community Organization':'people-group',
              "Veteran's Organization": 'people-roof',
              'Unsure / N/A': 'circle-question',
              'Other':'circle-question',
              'Food Truck': 'truck',
              'VFW': 'people-roof',
              'Market': 'store'
              }
records = df[['longitude', 'latitude','venue_name', 'venue_type','venue_address']].to_records()

containers_map = containers_map = folium.Map(location=[40.4406, -79.9959], zoom_start=11)
# Add the markers to the map...
for record in records:
    folium.Marker(location=[record[2], record[1]],
                  tooltip=record[3],
                  popup=record[5],
                  icon=folium.Icon(color=color_dict[record[4]],
                                   icon_color='white',
                                   icon=icon_dict[record[4]],
                                   prefix='fa',
                                   )
                  ).add_to(containers_map)


containers_map
```
